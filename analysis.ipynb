```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load all metrics files
metrics_files = [
    '/kaggle/input/analysis-data3/EfficientNetB4_Dataset1_metrics.csv',
    '/kaggle/input/analysis-data3/EfficientNetB4_Dataset3_metrics.csv',
    '/kaggle/input/analysis-data3/InceptionResNetV2_Dataset1_metrics.csv',
    '/kaggle/input/analysis-data3/InceptionResNetV2_Dataset3_metrics.csv',
    '/kaggle/input/analysis-data3/NASNetLarge_Dataset1_metrics.csv',
    '/kaggle/input/analysis-data3/NASNetLarge_Dataset3_metrics.csv'
]

all_metrics = pd.concat([pd.read_csv(f) for f in metrics_files])

# Set figure size and style
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid", palette="muted")

sns.barplot(x='Model', y='Test_Accuracy', hue='Dataset', data=all_metrics)
plt.title('Test Accuracy Comparison')
plt.ylim(0, 1)
plt.legend(loc='upper right')

# Save plot
plt.savefig('/kaggle/working/test_accuracy.eps', format='eps')

plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set figure size

sns.barplot(x='Model', y='Training_Time', hue='Dataset', data=all_metrics)
plt.title('Training Time (Seconds)')
plt.ylabel('Seconds')

plt.savefig('/kaggle/working/training_time.eps', format='eps')

plt.show()
```

```python
plt.figure(figsize=(12, 7))  # Bigger figure size

metrics_long = all_metrics.melt(id_vars=['Model', 'Dataset'], 
                               value_vars=['Training_Accuracy', 'Validation_Accuracy'],
                               var_name='Metric', 
                               value_name='Accuracy')

sns.lineplot(x='Model', y='Accuracy', hue='Dataset', style='Metric',
             markers=True, dashes=False, data=metrics_long, markersize=10)

plt.title('Training vs Validation Accuracy')
plt.ylim(0, 1)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.savefig('/kaggle/working/training_vs_validation_accuracy.eps', format='eps')

plt.show()
```
